```
# lib and packages
from IPython.core.debugger import set_trace
import pandas as pd
import numpy as np
import tensorflow as tf
import tensorflow_probability as tfp
import os
import random
import matplotlib.pyplot as plt
import pickle
import time
import math

%matplotlib inline
```

```
#generating values from Logit-Normal distrinution function
tfd = tfp.distributions
def tfd_logitnorm(num=1, mean=0, scale=1, seed=None): 
  tf.random.set_seed(seed)
  y = tfd.Sample(tfd.LogitNormal(mean,scale),num)
  return y.sample()

```

```
#generating mean and scale for all student
def mean_scale(num_stu, seed):
  random.seed(seed)
  mean_scale= {}
  for x in range(num_stu):
    mean = random.uniform(0.55,0.95)
    scale = random.uniform(0.15,0.4)
    mean_scale[x] = (mean,scale)
  return mean_scale
```

```
## student subject marks function
def student_marks(num_sub=1, num_stu=1, mean_scale={}, seed=None):
  stu_list = []
  for i in range(num_stu):
    stu1 = tfd_logitnorm(num_sub, mean_scale[i][0], mean_scale[i][1], seed)
    stu_list.append(stu1)
    all_stu_list = np.array(stu_list)
  return all_stu_list
```

```
#printing mean and scale values
avg_sd = mean_scale(num_stu=10, seed=1234)
avg_sd
```

```

#printing students marks sampled form Logit-Normal distribution
df3 = pd.DataFrame(student_marks(10, 10, avg_sd, seed=1234))
df3.columns = ['sub1','sub2','sub3','sub4','sub5','sub6','sub7','sub8','sub9','sub10']
df3
```
